```python
import pandas as pd
df_final_demo = pd.read_csv("C:/Users/Alexandra/Documents/Ironhack/Week5/Mini-Project/week5_project_ab_testing/data/raw/df_final_demo.txt", sep=',')
df_final_demo
```

```python
df_final_demo.dropna(how='all',inplace= True)
```

```python
df_final_demo.isnull().sum()
```

```python
df_final_demo.drop_duplicates(inplace=True)
```

```python
df_final_demo.dtypes
```

```python
df_final_demo["clnt_tenure_yr"]  = df_final_demo.clnt_tenure_yr.fillna('0')
df_final_demo["clnt_tenure_mnth"]  = df_final_demo.clnt_tenure_mnth.fillna('0')
df_final_demo["clnt_age"]  = df_final_demo.clnt_age.fillna('0')
df_final_demo["num_accts"]  = df_final_demo.num_accts.fillna('0')
df_final_demo["bal"]  = df_final_demo.bal.fillna('0')
df_final_demo["calls_6_mnth"]  = df_final_demo.calls_6_mnth.fillna('0')
df_final_demo["logons_6_mnth"]  = df_final_demo.logons_6_mnth.fillna('0')
```

```python
df_final_demo["clnt_tenure_yr"] = pd.to_numeric(
    df_final_demo["clnt_tenure_yr"],
    errors="coerce"
)
df_final_demo["clnt_tenure_mnth"] = pd.to_numeric(
    df_final_demo["clnt_tenure_mnth"],
    errors="coerce"
)

df_final_demo["num_accts"] = pd.to_numeric(
    df_final_demo["num_accts"],
    errors="coerce"
)

df_final_demo["calls_6_mnth"] = pd.to_numeric(
    df_final_demo["calls_6_mnth"],
    errors="coerce"
)


df_final_demo["logons_6_mnth"] = pd.to_numeric(
    df_final_demo["logons_6_mnth"],
    errors="coerce"
)
```

```python
#converting clnt_tenure_yr into an integer 
df_final_demo["clnt_tenure_yr"] = df_final_demo["clnt_tenure_yr"].astype("Int64")
df_final_demo["clnt_tenure_mnth"] = df_final_demo["clnt_tenure_mnth"].astype("Int64")
df_final_demo["num_accts"] = df_final_demo["num_accts"].astype("Int64")
df_final_demo["calls_6_mnth"] = df_final_demo["calls_6_mnth"].astype("Int64")
df_final_demo["logons_6_mnth"] = df_final_demo["logons_6_mnth"].astype("Int64")
```
