```python
df_demo.head()
```

```python

# Fill NaN with zeros otherwise it gives a error 
df_demo['clnt_tenure_yr'] = df_demo['clnt_tenure_yr'].fillna(0).astype(int)
print (df_demo['clnt_tenure_yr'])
```

```python
df_demo['clnt_tenure_mnth'] = df_demo['clnt_tenure_mnth'].fillna(0).astype(int)
print (df_demo['clnt_tenure_mnth'])
```

```python
df_demo['clnt_age'] = df_demo['clnt_age'].fillna(0).astype(int)
print (df_demo['clnt_age'])
```

```python
df_demo['calls_6_mnth'] = df_demo['calls_6_mnth'].fillna(0).astype(int)
print (df_demo['calls_6_mnth'])
```

```python
df_demo['logons_6_mnth'] = df_demo['logons_6_mnth'].fillna(0).astype(int)
print (df_demo['logons_6_mnth'])
```

```python
df_demo['num_accts'] = df_demo['num_accts'].fillna(0).astype(int)
print (df_demo['num_accts'])
```

```python
df_demo['gendr'].unique()
```

```python

# Define a categorizing method
gender_map = {
    'M': 'Male',
    'F': 'Female',
    'U': 'Unknown',
    'X': 'Non-binary', 
}


df_demo['gendr'] = df_demo['gendr'].replace(gender_map).fillna('Unknown')

print(df_demo)
```

```python
#generate csv file 
df_demo.to_csv('final_demo_cleaned.csv', index=False)  # i want to read csv 
```
