```python
#creating a new column called 'age_category' based on some limits from 'clnt_age'
young_age = df_analysis[df_analysis['clnt_age'] <= 19]
adult_age = df_analysis[(df_analysis['clnt_age'] >= 20) & (df_analysis['clnt_age'] <= 39)]
middle_adult_age = df_analysis[(df_analysis['clnt_age'] >= 40) & (df_analysis['clnt_age'] <= 64)]
old_age = df_analysis[df_analysis['clnt_age'] >= 65]
```

```python
df_analysis['age_category'] = np.select(
    [
        df_analysis['clnt_age'] <= 19,
        df_analysis['clnt_age'].between(20, 39),
        df_analysis['clnt_age'].between(40, 64),
        df_analysis['clnt_age'] >= 65
    ],
    ['Young', 'Adult', 'Middle Adult', 'Old'],
    default='Unknown'
)
```

```python
#checking the new column
df_analysis
```

```python
#seeing the number of clients from each category
df_analysis['age_category'].value_counts()
```

```python
#Piechart - General

counts = df_analysis['age_category'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(counts, autopct='%.1f%%', startangle=90, colors=sns.color_palette("Set3"), pctdistance=0.75)

plt.legend(counts.index, title="Age Category", loc="center left", bbox_to_anchor=(1, 0.5))

plt.ylabel("")
plt.show()
```

```python
#Checking the gender of people that are included in Middle adult category
df_analysis[df_analysis['age_category'] == 'Middle Adult']['gendr'].value_counts()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Get unique age categories
age_categories = df_analysis['age_category'].unique()

# Count males and females for each age category
genders = ['Male', 'Female', 'Unknown', 'Non-binary']

gender_counts = {
    'Male': np.array([df_analysis[(df_analysis['age_category'] == age) & 
                                  (df_analysis['gendr'] == 'Male')].shape[0]
                      for age in age_categories]),
    'Female': np.array([df_analysis[(df_analysis['age_category'] == age) & 
                                    (df_analysis['gendr'] == 'Female')].shape[0]
                        for age in age_categories]),
    'Unknown': np.array([df_analysis[(df_analysis['age_category'] == age) & 
                                    (df_analysis['gendr'] == 'Unknown')].shape[0]
                        for age in age_categories]),
    'Non-Binary':np.array ([df_analysis[(df_analysis['age_category'] == age) & 
                                    (df_analysis['gendr'] == 'Non-binary')].shape[0]
                        for age in age_categories])
}

width = 0.6  # width of the bars

# Plot
fig, ax = plt.subplots(figsize=(8,6))
bottom = np.zeros(len(age_categories))

for gender, counts in gender_counts.items():
    p = ax.bar(age_categories, counts, width, label=gender, bottom=bottom)
    bottom += counts
    ax.bar_label(p, label_type='center')

ax.set_xlabel('Age Category')
ax.set_ylabel('Count')
ax.set_title('Gender distribution by Age Category')
ax.legend()

plt.show()
```

```python
df_analysis['tenure_years']= df_analysis['clnt_tenure_mnth']/12
```

```python
demographic_analysis = (
    df_analysis
    .groupby('age_category')
    .agg(
        avg_balance=('bal', 'mean'),
        total_balance=('bal', 'sum'),
        total_logons=('logons_6_mnth', 'sum'),
        accounts=('num_accts', 'sum'),
        avg_tenure_years=('tenure_years', 'mean'),
        avg_logons=('logons_6_mnth','mean'),
        num_clients=('client_id', 'count')
    )
    .reset_index().sort_values(by = 'avg_balance', ascending=False)
)
```
