```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline

df_demo = pd.read_csv("../data/clean/demo_clean.csv", sep=",")
df_web = pd.read_csv("../data/clean/web_data_clean.csv", sep=",")
df_exp = pd.read_csv("../data/clean/experiment_clean.csv", sep=",")

df_demo, df_web, df_exp
```

```python
df_web_exp = df_web.merge(
    df_exp[["client_id", "variation"]],
    on="client_id",
    how="inner"
)
```

```python
df_web_exp
```

```python
df_web_exp["variation"].value_counts(dropna=False)
```

```python
df_web_exp = df_web_exp[df_web_exp["variation"].notna()]
df_web_exp
```

```python
df_copy = df_web_exp.copy()

online_clients = df_copy["client_id"].unique()
df_demo_online = df_demo[df_demo["client_id"].isin(online_clients)]
df_demo_online
```

```python
df_demo_online[["num_accts", "bal"]].describe()
```

```python
df_demo_online["gendr"].value_counts(normalize=True)
```

```python
df_demo_online["clnt_age"].describe()
```

```python
age_bins = [0, 30, 45, 60, 75, 100]
age_labels = ["<30", "30–44", "45–59", "60–74", "75+"]

df_demo_online["age_group"] = pd.cut(
    df_demo_online["clnt_age"],
    bins = age_bins,
    labels = age_labels
)

df_demo_online["age_group"].value_counts(normalize=True)
```

```python
df_demo_online["age_group"].value_counts().sort_index().plot(kind="bar")
plt.title("Age Distribution of Online Process Users")
plt.ylabel("Clients")
plt.show()
```
